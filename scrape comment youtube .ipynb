# n-grams
#Definir  des fonctions pour creer  n-grams

```python
#Definir la  fonction getMostCommon pour retourner et donner les mots communs     
from collections import Counter
def getMostCommon(reviews_list,topn=20):
    reviews=" ".join(reviews_list)
    tokenised_reviews=reviews.split(" ")
    
    
    freq_counter=Counter(tokenised_reviews)
    return freq_counter.most_common(topn) # retourner les mots avec les frequences elevÃ©s  
```

```python

#DÃ©finition d'une fonction plotMostCommonWords pour produire un barplot des mots les plus frequents
```

```python


def plotMostCommonWords(reviews_list,topn,title="Most Frequent Words",color="blue",axis=None): #coleur par defauts est bleue
    top_words=getMostCommon(reviews_list,topn=topn)
    data=pd.DataFrame()
    data['words']=[val[0] for val in top_words]
    data['freq']=[val[1] for val in top_words]
    if axis!=None:
        sns.barplot(y='words',x='freq',data=data,color=color,ax=axis).set_title(title+" top "+str(topn))
    else:
        sns.barplot(y='words',x='freq',data=data,color=color).set_title(title+" top "+str(topn))
```

```python
#Definir la fonction generateNGram pour creer  n-grams
def generateNGram(text,n):
    tokens=text.split(" ")
    ngrams = zip(*[tokens[i:] for i in range(n)])
    return ["_".join(ngram) for ngram in ngrams]
```

# 1 gram

```python
#Creer un 1 grams  pour  postifs  et  negatifs  commentaires  
goodcomment=[" ".join(generateNGram(review,1)) for review in goodreviews]
badcomment=[" ".join(generateNGram(review,1)) for review in badreviews]
```

```python

#En utilisant la fonction ci-dessus, pour produire des barplots pour les avis positifs et nÃ©gatifs sÃ©parÃ©ment
```

```python

from matplotlib import rcParams

rcParams['figure.figsize'] = 8,6 ## Sets the heigth and width of image


fig,ax=plt.subplots(1,2)
fig.subplots_adjust(wspace=0.5) #Adjusts the space between the two plots
plotMostCommonWords(goodcomment,20,"Positive comments Unigrams",axis=ax[0])

plotMostCommonWords(badcomment,20,"Negative comments Unigrams",color="red",axis=ax[1])
```

# bigrams 

```python
#Creer des bigrams  pour  postifs  et  negatifs  commentaires
goodcomment=[" ".join(generateNGram(review,2)) for review in goodreviews]
badcomment=[" ".join(generateNGram(review,2)) for review in badreviews]
```

```python

#En utilisant la fonction ci-dessus, pour produire des barplots pour les avis positifs et nÃ©gatifs sÃ©parÃ©ment
rcParams['figure.figsize'] = 6,8
fig,ax=plt.subplots(1,2)
fig.subplots_adjust(wspace=1) #Adjusts the space between the two plots
plotMostCommonWords(goodcomment,25,"Positive comments Bigrams",axis=ax[0])
# plotting both most frequently positive and negative trigrams
plotMostCommonWords(badcomment,25,"Negative comments Bigrams",color="red",axis=ax[1])
```
